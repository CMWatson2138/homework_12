### Import necessary modules and retrieve dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

file_to_load = 'loans_full_schema.csv'
lc_df = pd.read_csv(file_to_load)
pd.options.display.max_columns=70
lc_df
```

### The first step in preprocessing is to clean up null values.  In order to do that I will first determine which columns contain null values and how they should be handled.

```python
lc_df.info()
```

### Columns containing null values are as follows [emp_title, emp_length, debt_to_income, annual_income_joint, verification_income_joint, debt_to_income_joint, months_since_last_delinq, months_since_90d_late, months_since_last_credit_inquiry, num_accounts_120d_past_due]. The first column I replaced NaN with Not Provided.  For emp_length it doesn't make sense to insert a number, as all have verified income, so the number cannot be 0.  For debt_to_income this is a relatively small number of accounts so we should drop these rows.  All three values that use joint accounts should be left alone but any calculations should check for values here before moving on to individual income info. Each of the NaN values for delinquency and past due, as well as months since inquiry should be replaced with 0, as these are not applicable because they are accounts with liens or bankruptcies and that information is captured elsewhere.

```python
# Placed Not Provided in emp_title with NaN
lc_df["emp_title"].fillna("Not Provided", inplace = True)
# Removed values where debt_to_income was NaN
#lc_df['debt_to_income'].dropna(axis=0, how='any', inplace=True)
# Filled each NaN with 0 as these are not applicable fields
lc_df['months_since_last_delinq'].fillna(0, inplace=True)
lc_df['months_since_90d_late'].fillna(0, inplace=True)
lc_df['months_since_last_credit_inquiry'].fillna(0, inplace=True)
lc_df['num_accounts_120d_past_due'].fillna(0, inplace=True)
```

### The next step is data cleaning.  In this step I will change the issue_month column to reflect DateTime.

```python
lc_df['issue_month'] = pd.to_datetime(lc_df['issue_month'], format='%b-%Y', errors='ignore')
```

### This step is to handle missing data, however, I did that in the first step with NaN changes.

### This step will remove extreme outliers.  I chose to remove outliers with IQR method in annual_income and total_credit_limit, as that was where I saw large skew

```python
#Looking at what the data is comprised of.  Surpressed scientific notation with lambda
lc_df.describe().apply(lambda s: s.apply('{0:.5f}'.format))
```

```python
mod_lc_df= lc_df.copy()
#Calculated the IQR for annual_income
Q1 = mod_lc_df['annual_income'].quantile(0.25)
Q3 = mod_lc_df['annual_income'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Calculating the upper and lower bounds of acceptable outliers for annual_income
upper = np.where(mod_lc_df['annual_income'] >= (Q3+1.5*IQR))
lower = np.where(mod_lc_df['annual_income'] <= (Q1-1.5*IQR))
#Removing those outliers
mod_lc_df.drop(upper[0], inplace = True)
mod_lc_df.drop(lower[0], inplace = True)
#review results
mod_lc_df.describe().apply(lambda s: s.apply('{0:.5f}'.format))
```

```python
#Calculated the IQR for total_credit_limit
Q1 = mod_lc_df['total_credit_limit'].quantile(0.25)
Q3 = mod_lc_df['total_credit_limit'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Calculating the upper and lower bounds of acceptable outliers for total_credit_limit
upper1 = np.where(mod_lc_df['total_credit_limit'] >= (Q3+1.5*IQR))
lower1 = np.where(mod_lc_df['total_credit_limit'] <= (Q1-1.5*IQR))
#Removing those outliers
mod_lc_df.drop(upper1[0], inplace = True, errors='ignore')
mod_lc_df.drop(lower1[0], inplace = True, errors='ignore')
#review results
mod_lc_df.describe().apply(lambda s: s.apply('{0:.5f}'.format))
```

### I believe the function defined below is a better way to do this than lambda.  It seems more concise, and it will be easier to use again if we receive more data.

## b.	Either use two different lambdas or a lambda and a different technique and see which one is more performant. Explain your findings in the markdown. 


```python
region=[]
def region_mapper():
    for st in lc_df['state']:
        if st in Pacific:
            region.append('Pacific')
        elif st in Plains:
            region.append('Plains')
        elif st in Midwest:
            region.append('Midwest')
        elif st in Northeast:
            region.append('Northeast')
        elif st in South:
            region.append('South')
        else:
            region.append('unknown')
    lc_df['region'] = region
    
region_mapper()
lc_df
```
